```python
import tensorflow as tf
import os
import random
from scipy import misc
import numpy as np
source_file="D:/img/42/"       #原始文件地址
target_file="D:/img/resize/42/"  #修改后的文件地址
if not os.path.exists(target_file):  
    os.makedirs(target_file)
file_list=os.listdir(source_file)   #读取原始文件的路径
with tf.Session() as sess:
    for i in range(len(file_list)):
        image_raw_data=tf.gfile.FastGFile(source_file+file_list[i],"rb").read()#读取图片
        print("正在处理：",file_list[i])
        image_data=tf.image.decode_jpeg(image_raw_data)
        img_resized = tf.image.resize_images(image_data, [512, 512], method=0)
        img_resized = np.asarray(img_resized.eval(), dtype='uint8')
        grayed_image = tf.image.rgb_to_grayscale(img_resized)
        encode_data=tf.image.encode_jpeg(img_resized) #图片颜色
        with tf.gfile.GFile(target_file+str(i)+".jpg","wb") as f:
            f.write(encode_data.eval())                        
print("图像处理完毕")
```
